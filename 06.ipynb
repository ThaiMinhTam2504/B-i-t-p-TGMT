```python
#import thư viện
from PIL import Image
from imgtools import *
```

```python
#đường dẫn ảnh
my_path= r'C:/Users/ADMIN/Pictures/MEME/doremon.jpg'
```

```python
#đọc ảnh
img=load_image(my_path)
```

```python
display(img)
```

```python
#chuyển từ ảnh rgb sang ảnh xám
gray_image=img.convert("L")
display(gray_image)
```

```python
#phân rgb thành 3 màu khác nhau
red_band,green_band,blue_band=img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
#trộn ảnh
merged_image=Image.merge("RGB",(red_band,green_band,blue_band))
display(merged_image)
```
