```python
#Import thư viện
from PIL import Image
```

```python
#Thư mục chứa hình ảnh
dir ='C:/Users/ADMIN/Pictures/MEME'
```

```python
#Đường dẫn đến ảnh
image_path = dir + '/doremon.jpg'
print(image_path)
```

```python
#sử dụng Image.open() để đọc ảnh
img01=Image.open(image_path)
```

```python
#hiển thị hình ảnh
display(img01)
```

```python
# In một số thông tin
# In định dạng ảnh
print("Định dang ảnh: ", img01.format)
```

```python
#in ra size
print("Kích thước ảnh: ",img01.size)
```

```python
#Lưu ảnh và chuyển đổi ảnh
new_img_01_path=dir + '/new_01.JPG'
new_img_png_01_path=dir+ '/new_01.PNG'
img01.save(new_img_01_path)
img01.save(new_img_png_01_path)
```

```python
#đóng tập tin
img01.close()
```
